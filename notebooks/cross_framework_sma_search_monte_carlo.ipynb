# Cross Framework SMA Search And Monte Carlo

This notebook calls the quant-orchestrator job that runs vectorbt search, Monte Carlo, and forward tests across fmp and yfinance.


```python
from pathlib import Path
import sys

repo_root = Path.cwd().resolve()
if not (repo_root / 'quant_orchestrator').exists():
    repo_root = repo_root.parent
if str(repo_root) not in sys.path:
    sys.path.insert(0, str(repo_root))

from quant_orchestrator.backtests.research import run_cross_framework_sma_search_monte_carlo


result = run_cross_framework_sma_search_monte_carlo(
    providers=('fmp', 'yfinance'),
    forward_symbols=('QQQ', 'SPY'),
    capital_base=100_000.0,
    top_mc_count=10,
    mc_iterations=1_000,
    mc_block_size=5,
    start='2020-01-01',
    train_end='2025-12-31',
    test_start='2026-01-01',
    test_end='2026-12-31',
)

print('Vectorized search:')
print(result['search_elapsed'].to_string(index=False))
print(result['train_search_report'].loc[:, ['provider', 'fast_window', 'slow_window', 'final_equity', 'total_return', 'max_drawdown', 'train_rank']].to_string(index=False))
print()
print('Monte Carlo:')
print(result['mc_elapsed'].to_string(index=False))
print(result['mc_report'].to_string(index=False))
print()
print('Selections:')
print(result['selections'].to_string(index=False))
print()
print('Forward test:')
print(result['forward_report'].loc[:, ['provider', 'selection', 'framework', 'symbols', 'fast_window', 'slow_window', 'final_equity', 'total_return', 'max_drawdown', 'elapsed_seconds']].to_string(index=False))
print()
print('Job summary:')
print(result['summary'].to_string(index=False))
```
