# Multi Vendor Backtesting.py SMA Crossover

This notebook calls the quant-orchestrator job that compares fmp and yfinance using a fixed backtesting.py SMA crossover.


```python
from quant_orchestrator.backtests.research import run_multi_vendor_backtesting_py_sma_comparison


result = run_multi_vendor_backtesting_py_sma_comparison(
    providers=('fmp', 'yfinance'),
    start='2020-01-01',
    end=None,
    fast_window=50,
    slow_window=200,
    capital_base=100_000.0,
)
symbol_report, portfolio_report, runs_by_provider = result

print('Symbol-level report:')
print(symbol_report.loc[:, ['provider', 'symbol', 'start', 'end', 'bars', 'trades', 'final_equity', 'total_return', 'max_drawdown', 'elapsed_seconds']].to_string(index=False))
print()
print('Portfolio report:')
print(portfolio_report.loc[:, ['provider', 'symbol', 'bars', 'trades', 'final_equity', 'total_return', 'max_drawdown', 'elapsed_seconds']].to_string(index=False))
print()
comparison = portfolio_report.loc[:, ['provider', 'total_return', 'max_drawdown', 'daily_vol', 'elapsed_seconds']].set_index('provider')
comparison['return_diff_vs_fmp'] = comparison['total_return'] - float(comparison.loc['fmp', 'total_return'])
comparison['drawdown_diff_vs_fmp'] = comparison['max_drawdown'] - float(comparison.loc['fmp', 'max_drawdown'])
print('Portfolio comparison:')
print(comparison.to_string())
```
